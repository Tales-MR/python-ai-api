```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import scipy.stats as stats
import seaborn as sns
```

```python
dataset_base = pd.read_csv("./dataset/mtcars.csv")


data = dataset_base.drop(['model'], axis=1)
```

```python
corr = data.corr()

sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
```

```python
column_pairs = [('mpg','cyl'),('mpg','disp'),('mpg','hp'),('mpg','drat'),('mpg','wt'),('mpg','vs')]

n_plots = len(column_pairs)

fig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(6, 4 * n_plots))

for i, pair in enumerate(column_pairs):
    x_col, y_col = pair
    
    sns.scatterplot(x=x_col, y=y_col, data=data, ax=axes[i])
    axes[i].set_title(f'{x_col} vs {y_col}')
    
plt.tight_layout()
plt.show()
```

```python
modelo = sm.ols(formula='mpg ~ cyl + disp + hp + wt', data=data)
modelo = modelo.fit()
modelo.summary()
```

```python
modelo.predict([4, 121, 109, 2.78])
```
